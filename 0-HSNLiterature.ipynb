# Human Single Neuron Literature

```python
import numpy as np

from lisc import Counts1D, Counts, Words
from lisc.collect import collect_across_time

from lisc.plts.counts import plot_matrix
from lisc.plts.time import plot_results_across_years
from lisc.plts.utils import save_figure
```

```python
import seaborn as sns
sns.set_context('talk')
```

## Terms & Settings

```python
# Define main search terms
hsn_terms = [
    ['single-neuron', 'single-unit']
]

# Define inclusion terms - terms which are required for a paper to be included
inclusions = [
    ['human', 'patient']
]

# Define exclusion terms - terms which if present reject a paper from being included
exclusions = [
    ['cat', 'rabbit', 'rodent', 'rat', 'mouse', 'monkey', 'macaque', 'rhesus', 'primate', 'chimpanzee',
     'worm', 'elegans', 'retina', 'stem cells', 'organoid', 'muscle',
     'blood', 'platelet', 'plasma', 'protein', 'proteins', 'PCR', 'enzyme',
     'kidney', 'gastrointestinal', 'intestine', 'liver', 'lymph', 'urinary', 'renal', 'esophageal',
     'infection', 'virus', 'vaccine', 'vitamin', 'metabolism', 'RNA', 'DNA',
     'pregnancy', 'neonates', 'neonatal', 'oncoloy', 'cancer', 'carcinoma', 'insulin']
]
```

```python
# Define a range of years to collect data for
years = np.arange(1960, 2030, 5)
```

```python
# Save settings
SAVE_FIG = True
```

## Counts

```python
# Initialize the counts object
counts = Counts1D()
```

```python
# Add search terms to object
counts.add_terms(hsn_terms)
counts.add_terms(inclusions, 'inclusions')
counts.add_terms(exclusions, 'exclusions')
```

```python
# Collect counts data across years
years_counts = collect_across_time(counts, years)
```

```python
# Plot the results over time
plot_results_across_years(years_counts, figsize=(5, 3))
if SAVE_FIG:
    save_figure('hsn_years.pdf', 'figs', bbox_inches='tight')
```

## Words

```python
# Initialize words collection object
words = Words()
```

```python
# Add search terms to object
words.add_terms(hsn_terms)
words.add_terms(inclusions, 'inclusions')
words.add_terms(exclusions, 'exclusions')
```

```python
# Collect words data across years
years_words = collect_across_time(words, years, retmax=250)
```

```python
# Check the articles per time period
for time_period, cwords in years_words.items():
    print('---- {} ----'.format(time_period))
    cwords.check_articles()
```
